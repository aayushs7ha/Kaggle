**Boston Crime Dataset is a collection of crime incident reports that are provided by the boston police department (BPD).**
- This dataset documents the initial details surrounding an incident to which BPD officers respond
- This is a dataset containing records from the new crime incident report system, which includes a reduced set of fields focused on capturing the type of incident as well as when and where it occured.
- Records in the new system begin from June 2015

# Data Dictionary
- INCIDENT_NUMBER -> Internal BPD report number 
- OFFENSE_CODE -> Numerical code of offense description
- OFFENSE_CODE_GROUP -> Internal categorization of offense description
- OFFENSE_DESCRIPTION -> The primary description of the incident
- DISTRICT - what district crime was reported in
- REPORTING_AREA -> RA number associated with the location where the crime was reported from
- SHOOTING ->  Indicated, a shooting took place
- OCCURRED_ON_DATE -> Earliest date and time the incident could have taken place
- UCR_Part -> Universal Crime Reporting Part Number (1,2,3)
- STREET -> Street name the incident took place
- Lat -> Latitude
- Long -> Longitude
- Location -> Location

```python
crime_data = pd.read_csv("/kaggle/input/crimes-in-boston/crime.csv",encoding='windows-1254')
offense_code = pd.read_csv("/kaggle/input/crimes-in-boston/offense_codes.csv",encoding='windows-1254')
```

# Merge Dataset 


```python
bostonCrime = crime_data.merge(offense_code,left_on='OFFENSE_CODE',right_on='CODE',how='left').drop('CODE',axis='columns')

```
